Un perceptron es una red neuronal con solo una neurona.

```python
import numpy as np
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
```

```python
def sigmoid_derivative(x):
  return x*(1-x)
```

```python
training_inputs = np.array([[0,0,1],
                            [1,1,1],
                            [1,0,1],])
shape = training_inputs.shape
```

```python
training_outputs = np.array([[0,1,1]]).T
```

```python
synaptic_weights = np.random.randn(shape[1], 1) * np.sqrt(1/shape[1]) #mejora respecto al vídeo
synaptic_weights
```

```python
for iteration in range(20000):
  input_layer = training_inputs
  outputs = sigmoid(np.dot(input_layer, synaptic_weights))
  error = training_outputs - outputs
  adjustments = error * sigmoid_derivative(outputs)
  synaptic_weights += np.dot(input_layer.T, adjustments)

print(synaptic_weights)
print(outputs)
print(np.abs(error))
```

```python
test_inputs = np.array([[0,1,1]])
test_outputs = np.array([[0]]).T #como el primer elemeto es un 0, el output debe ser 0
```

```python
input_layer = test_inputs
outputs = sigmoid(np.dot(input_layer, synaptic_weights))
error = test_outputs - outputs
print(outputs)
print(np.abs(error))
```
